# Stacking
joining of two numpy array.

```python
# Horizontal Stacking.
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
a5
```

```python
np.hstack((a4,a5))
```

```python
np.hstack((a4,a5,a4,a5))
```

```python
# Vertical Stacking
np.vstack((a4,a5))
```

```python
display(np.vstack((a4,a5)))
```
