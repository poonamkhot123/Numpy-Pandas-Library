# Group by = Categorical data

```python
moives = pd.read_csv("/content/imdb-top-1000.csv")
```

```python
genres = moives.groupby("Genre")
```

```python
# Applying builtin aggregation functions on groupby objects.
genres.max()
```

```python
# find the top 3 genres by total earning.
moives.groupby("Genre").sum()["Gross"].sort_values(ascending=False).head(3)
```

```python
moives.groupby("Genre")["Gross"].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with highest avg IMDB rating.
moives.groupby("Genre")["IMDB_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity.
moives.groupby("Director")["No_of_Votes"].sum().sort_values(ascending=False).head(1)
```

```python
ipl = pd.read_csv("/content/deliveries.csv")
ipl.head()
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs scored.
ipl.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False)
```

```python
# find the batsman with max no of sixes.
six = ipl[ipl["batsman_runs"]==6]
six.groupby("batsman")["batsman"].count().sort_values(ascending=False)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs.
temp_df=ipl[ipl["over"]>15]
temp_df[(temp_df["batsman_runs"]==4) | (temp_df["batsman_runs"] == 6)]
temp_df.groupby("batsman")["batsman"].count().sort_values(ascending=False)
```

```python
# find v kohli's record against all teams.
ipl[ipl["batsman"]=="V Kohli"].groupby("bowling_team")["batsman_runs"].sum()
```
