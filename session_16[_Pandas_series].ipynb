# what is pandas

# Pandas series
* A pandas Series is like a column in a table. it is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

# Series from lists.

```python
# string
country = ['india','pakistan','USA','shrilanka']
pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]
pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a name
pd.Series(marks,index=subjects,name="Poonam's marks")
```

```python
marks=pd.Series(marks,index=subjects,name="Poonam's marks")
marks
```

# Series from dict
